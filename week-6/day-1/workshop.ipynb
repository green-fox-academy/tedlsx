```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])
most_recent_ten = df.sort_values(by = "Last Updated", ascending=False)[:10]
most_recent_ten
```

```python
top = int(0.25 * df.count()["Rating"])

most_recent = most_recent_ten["Last Updated"][0]

# most_two_recent_mon = pd.period_range(most_recent, periods=2, freq="M")
most_two_recent_mon = [most_recent.month, most_recent.month - 1]
most_year = most_recent.year

df_top = df.sort_values(by = "Rating", ascending=False)[:top]
# df_top["Last Updated"].dt.year.isin([2018,2017])
df_top[(df_top["Last Updated"].dt.year == most_year) & (df_top["Last Updated"].dt.month.isin(most_two_recent_mon))]
```

```python
df["Size"].unique()
```

```python
def find_size(row):
    if "k" in row:
        return float(row[:-1]) / 1024
    elif row == "Varies with device":
        return 0
    else:
        return float(row[:-1])
    
df_size = df["Size"].apply(find_size)
max_size = df_size.sort_values(ascending=False)[0]
size_list = df_size[df_size==max_size].index.values.tolist()
df[df.index.isin(size_list)].sort_values(by="Rating", ascending=True)
```

```python
df.loc[:, "new_reviews"] = df.apply(lambda row: f'Rating is {row["Rating"]} and Reviews counts is {row["Reviews"]}', axis = 1)
```
